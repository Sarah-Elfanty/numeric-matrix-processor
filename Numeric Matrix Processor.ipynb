```python
def mat_add(matrixA, matrixB):
    
    output = []
    
    for rowA,rowB in zip(matrixA, matrixB):
        rows = []
        for elementA, elementB in zip(rowA, rowB):
            rows.append(elementA + elementB)   
        
        output.append(rows)
    
    return output     

def const_mul(matrix, const):
    
    output = []
    
    for row in matrix:
        rows = []
        for element in row:
            rows.append(element * const)
            
        output.append(rows)
        
    return output
    
def mat_mul(matrixA, matrixB):
    
    output = []
       
    
    for i in range(len(matrixA)):
        row = []
        for k in range(len(matrixB[0])):
            dot_prod = 0
            for j in range(len(matrixA[i])):
                dot_prod += matrixA[i][j] * matrixB[j][k]
            row.append(dot_prod)
        output.append(row)
    
    return output

def mat_transpose():
    print("""1. Main diagonal
2. Side diagonal
3. Vertical line
4. Horizontal line""")
    
    choice = int(input("Your choice: "))
    
    if choice == 1:
        size = input("Enter size of matrix: ").split()
        print("Enter matrix:")
        matrix = get_matrix(size)
        print_matrix(transpose_main(matrix, size))
    elif choice == 2:
        size = input("Enter size of matrix: ").split()
        print("Enter matrix:")
        matrix = get_matrix(size)
        print_matrix(transpose_side(matrix , size))
    elif choice == 3:
        size = input("Enter size of matrix: ").split()
        print("Enter matrix:")
        matrix = get_matrix(size)
        print_matrix(transpose_vert(matrix, size))
    else:
        size = input("Enter size of matrix: ").split()
        print("Enter matrix:")
        matrix = get_matrix(size)
        print_matrix(transpose_horiz(matrix, size))

def transpose_main(matrix, size):
    
    output = []
    for j in range(int(size[1])):
        rows = []
        for i in range(int(size[0])):
            
            rows.append(matrix[i][j])
        
        output.append(rows)
        
    return output

def transpose_side(matrix, size):
    
    output = []
    for j in range(int(size[1]) - 1, -1, -1):
        rows = []
        for i in range(int(size[0]) - 1, -1, -1):
            rows.append(matrix[i][j])
            
        output.append(rows)
        
    return output

def transpose_vert(matrix, size):
    
    output = []
    for i in range(int(size[0])):
        rows = []
        for j in range(int(size[1]) - 1, -1, -1):
            rows.append(matrix[i][j])
            
        output.append(rows)
        
    return output

def transpose_horiz(matrix, size):
    
    output = []
    for i in range(int(size[0]) - 1, -1, -1):
        rows = []
        for j in range(int(size[1])):
            rows.append(matrix[i][j])
            
        output.append(rows)
        
    return output
```

```python
def cal_det(matrix):
    
    if len(matrix) < 2 :
        det = matrix[0][0]
    elif len(matrix) == 2:
        det = (matrix[0][0] * matrix[1][1]) - (matrix[0][1] * matrix[1][0])
    else:
        det = 0

        for j, element in enumerate(matrix[0]):
            
            minor = []
            for row in (matrix[1:]):
                minor_elements = []
                for k, Element in enumerate(row):
                    if k != j:
                        minor_elements.append(Element)
                minor.append(minor_elements)

            cofactor = ((-1) ** (j + 2)) * cal_det(minor)
            
            det += element * cofactor
    return det
```

```python
def get_cofactor(matrix):
    
    
    cofactors_matrix = []
    for i in range(len(matrix)):
        cofactor = []
        for j in range(len(matrix[i])):
            
            minor = []
            for l, row in enumerate(matrix[0:]):
                minor_elements = []
                for k, Element in enumerate(row):
                    
                    if k != j and l != i:
                        minor_elements.append(Element)
                if len(minor_elements) != 0:
                    minor.append(minor_elements)
                    
            cofactor.append(((-1) ** (i + j + 2)) * cal_det(minor))
        cofactors_matrix.append(cofactor)
            
    return cofactors_matrix
```

```python
def get_matrix(size):
    matrix = []
    
    for i in range(int(size[0])):
        matrix.append(list(map(eval, input().split())))
    return matrix

def print_matrix(matrix):
    print("The result is:")
    for row in matrix:
        for element in row:
            print(element, end=' ')
        print()
    print()
```

```python
def get_inverse(matrix, size, det):
    
    return const_mul(transpose_main(get_cofactor(matrix), size), (1 / det)) 
```
